```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 30
train_loss = np.random.uniform(size=n)
eval_loss = np.random.uniform(size=n)
train_acc = np.random.uniform(size=n)
eval_acc = np.random.uniform(size=n)
```

```python
data = pd.DataFrame(
    data={"epoch": np.arange(n), "train loss": train_loss, "evaluation loss": eval_loss, "train accuracy": train_acc, "evaluation accuracy": eval_acc})
data
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(20,10))
data.plot(x='epoch', y='train loss', color='C1', ax=ax1)
data.plot(x='epoch', y='evaluation loss', color='C2', ax=ax1)
data.plot(x='epoch', y='train accuracy', color='C3', ax=ax2)
data.plot(x='epoch', y='evaluation accuracy', color='C4', ax=ax2)
plt.show()
```
