```python
import os
from astropy.table import Table
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
root_path = '/home/mingx/datasets/ground_extract/GroundBasedTraining'
data0 = os.path.join(root_path, 'Data.0/Public')
data1 = os.path.join(root_path, 'Data.1/Public')
print(data0, data1)
```

```python
band = ['R', 'I', 'G', 'U']
data0_band1 = os.listdir(os.path.join(data0, 'Band1'))
print(len(data0_band1), data0_band1[0])
```

```python
start, dist = 100, 10
files = data0_band1
plt.rcParams['figure.figsize'] = [15, 15]
for z in range(dist):
    i = start + z
    rfile = files[i]
    num = rfile[12:18]
    bands = [f"Band1/{rfile}"]
    for j in range(2, 5):
        bands.append(f"Band{j}/imageSDSS_{band[j-1]}-{num}.fits")
    for k in range(len(bands)):
        fpath = os.path.join(data0, bands[k])
        img = fits.getdata(fpath)
        img *= 256
        ax = plt.subplot(dist, 4, z * 4 + k + 1)
        plt.imshow(img)
```
